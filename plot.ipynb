```python
import matplotlib.pyplot as plt
import json

colors = ['r', 'g', 'b', 'c', 'y']

for _idx in range(5):
    with open(f'record{_idx}.json', 'r') as fp:
        record = json.load(fp)
    lg = []
    plt.title(f'Model accuracy: Bot {_idx}')
    cp = 0
    for name, mx in [('mainbody', 3), ('bywings', 2)]:
        for k in range(mx):
            h = record[f'{name}{k}']
            plt.plot(h[1], linestyle='--', color=colors[cp])
            plt.plot(h[3], color=colors[cp])
            cp += 1
            plt.ylabel('Accuracy')
            plt.xlabel('Epoch')
            lg.append(f'Train {name} {k}')
            lg.append(f'Validate {name} {k}')
    plt.legend(lg, loc='upper left')
    plt.show()

    plt.title(f'Model loss: Bot {_idx}')
    cp = 0
    for name, mx in [('mainbody', 3), ('bywings', 2)]:
        for k in range(mx):
            h = record[f'{name}{k}']
            plt.plot(h[0], linestyle='--', color=colors[cp])
            plt.plot(h[2], color=colors[cp])
            cp += 1
            plt.ylabel('Loss')
            plt.xlabel('Epoch')
    plt.legend(lg, loc='upper left')
    plt.show()
```
